```python
#As the number is different some issue is there, so we will again construct a loop but 
#this time considering one block of Laptop features at a time and to deal with not same length of elements in 
#each column ,will use try and except .
```

```python
# Create empty lists to store the extracted information.
names = []
ratings = []
reviews = []
prices = []
display = []
OS = []
RAM = []
processor=[]
SSD_capacity=[]
Warranty_period=[]
MRP=[]
Off=[]
```

```python
# Find all the laptop items on the webpage.
laptops = soup.find_all("div", {"class": "_2kHMtA"})
#laptops
```

```python
for i in laptops:
    #Extract the Name of the Laptop.
    try:
        name=i.find("div",{"class":"_4rR01T"}).text
    except:
        name=None
    names.append(name)
    
    
    # Extract the rating of the laptop
    try:
        rating = i.find("div", {"class":"_3LWZlK"}).text
    except :
        rating = None
    ratings.append(rating)
    
    
    #Extract the Reviews of the Laptop
    try:
        review=i.find("span",{"class":"_2_R_DZ"}).text
    except:
        review=None
    reviews.append(review)
     
    #Extract the Price of the Laptop
    try:
        cost=i.find("div",{"class":"_30jeq3"}).text
    except:
        cost=None
    prices.append(cost)
    
    #Extract the Display of the Laptop
    try :
        dp = i.findAll("li",{"class":"rgWa7D"})[4].text
    except:
        dp = None 
    display.append(dp)
    
    
    #Extract the Processor of the Laptop
    try:
        pro=i.findAll("li",{"class":"rgWa7D"})[0].text
    except:
        pro=None
    processor.append(pro)
    
    
    #Extract the RAM of the Laptop
    try:
        ram=i.findAll("li",{"class":"rgWa7D"})[1].text
    except:
        ram=None
    RAM.append(ram)

    
    #Extract the Operatin System of the Laptop
    try:
        os=i.findAll("li",{"class":"rgWa7D"})[2].text
    except:
        os=None
    OS.append(os)
    
    #Extract the SSD_Capacity of the Laptop
    try:
        space=i.findAll("li",{"class":"rgWa7D"})[3].text
    except:
        sapce=None
    SSD_capacity.append(space)
    
    
    #Extract the Warranty year of the Laptop
    try:
        wy=i.findAll("li",{"class":"rgWa7D"})[-1].text
    except:
        wy=None
    Warranty_period.append(wy)
    
    #Extract the MRP of the Laptop
    try:
        mrp=i.find("div",{"class":"_3I9_wc"}).text
    except:
        mrp=None
    MRP.append(mrp)
    
    
    #Extract the Off On Laptop
    try:
        off=i.find("div",{"class":"_3Ay6Sb"}).text
    except:
        off=None
    Off.append(off)
    
    
    
    
    
```

```python
df=pd.DataFrame({'Laptop_name':names,
                 'Rating':ratings,
                 'Review':reviews,
                 'MRP':MRP,
                 'Flipkart_Price':prices,
                 'Off_%':Off,
                 'Display_size':display,
                 'Processor':processor,
                 'RAM':ram,
                 'OS':os,
                 'SSD_Capacity':SSD_capacity,
                 'Warranty_period':Warranty_period})
df
```

```python
for j in range(1,10):
    url = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(j)
    response = requests.get(url)
    response
    
    webpage = response.text
    webpage
    
    soup=BeautifulSoup(webpage,"html.parser")
    
    
    names = []
    ratings = []
    reviews = []
    prices = []
    display = []
    OS = []
    RAM = []
    processor=[]
    SSD_capacity=[]
    Warranty_period=[]
    MRP=[]
    Off=[]
    
    
    laptops = soup.find_all("div", {"class": "_2kHMtA"})
    
    for i in laptops:
        #Extract the Name of the Laptop
        try:
            name=i.find("div",{"class":"_4rR01T"}).text
        except:
            name=None
        names.append(name)
    
    
        # Extract the rating of the laptop
        try:
            rating = i.find("div", {"class":"_3LWZlK"}).text
        except :
            rating = None
        ratings.append(rating)
    
    
        #Extract the Reviews of the Laptop
        try:
            review=i.find("span",{"class":"_2_R_DZ"}).text
        except:
            review=None
        reviews.append(review)
     
        #Extract the Price of the Laptop
        try:
            cost=i.find("div",{"class":"_30jeq3"}).text
        except:
            cost=None
        prices.append(cost)
    
        #Extract the Display of the Laptop
        try :
            dp = i.findAll("li",{"class":"rgWa7D"})[4].text
        except:
            dp = None 
        display.append(dp)
    
    
        #Extract the Processor of the Laptop
        try:
            pro=i.findAll("li",{"class":"rgWa7D"})[0].text
        except:
            pro=None
        processor.append(pro)
    
    
        #Extract the RAM of the Laptop
        try:
            ram=i.findAll("li",{"class":"rgWa7D"})[1].text
        except:
            ram=None
        RAM.append(ram)

    
        #Extract the Operatin System of the Laptop
        try:
            os=i.findAll("li",{"class":"rgWa7D"})[2].text
        except:
            os=None
        OS.append(os)
    
        #Extract the SSD_Capacity of the Laptop
        try:
            space=i.findAll("li",{"class":"rgWa7D"})[3].text
        except:
            sapce=None
        SSD_capacity.append(space)
    
    
        #Extract the Warranty year of the Laptop
        try:
            wy=i.findAll("li",{"class":"rgWa7D"})[-1].text
        except:
            wy=None
        Warranty_period.append(wy)
    
        #Extract the MRP of the Laptop
        try:
            mrp=i.find("div",{"class":"_3I9_wc"}).text
        except:
            mrp=None
        MRP.append(mrp)
    
    
        #Extract the Off On Laptop
        try:
            off=i.find("div",{"class":"_3Ay6Sb"}).text
        except:
            off=None
        Off.append(off)
    
    
    df=pd.DataFrame({'Laptop_specification':names,
                 'Rating':ratings,
                 'Review':reviews,
                 'MRP':MRP,
                 'Flipkart_Price':prices,
                 'Discount_%':Off,
                 'Display_size':display,
                 'Processor':processor,
                 'RAM':ram,
                 'OS':os,
                 'SSD_Capacity':SSD_capacity,
                 'Warranty_period':Warranty_period})
    
    final=pd.concat([final, df], ignore_index=True)
```
