```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Uncleaned_Flipkart_LAPTOP_dataset.csv")
```

```python
data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
data.head()
```

```python
# function to extract the substring before "Laptop"
def extract_name(Laptop_name):
    name = Laptop_name.split("Core")[0]
    return name
```

```python
# apply the function to the entire column using the apply() method
# insert the Laptop Name column next to the first column
data.insert(1, 'Laptop Name', data['Laptop_specification'].apply(extract_name))
```

**CLEANING THE MESS**

```python
data.loc[[2,123,141,152,200],['Display_size']]=data.loc[[2,123,141,152,200],['SSD_Capacity']].values
data.loc[[2,123,141,152,200],['SSD_Capacity']]=None
```

```python
data.loc[[32],['SSD_Capacity']]=data.loc[[32],['Display_size']].values
data.loc[[32],['Display_size']]=None
```

```python
data.loc[[74,110,208],['Display_size']]=None
data.loc[[74,110,208],['SSD_Capacity']]=None
```

```python
data.loc[[178],['SSD_Capacity']]=data.loc[[178],['Display_size']].values
data.loc[[178],['Display_size']]=None
```

```python
data.loc[32]
```

```python
# remove the Indian Rupee symbol and commas from price and MRP columns
data["MRP"] = data["MRP"].str.replace("₹", "").str.replace(",", "").astype(float)
data["Flipkart_Price"] = data["Flipkart_Price"].str.replace("₹", "").str.replace(",", "").astype(float)
```

```python
# extract percentage discount from the "Off_%" column and rename it to "Discount_%"
data["Discount_%"] = data["Discount_%"].str.replace("% off", "").astype(float)
```

```python
data['SSD_Capacity'].unique()
```

```python
capacity_map = {'512 GB SSD': 512,
                '256 GB SSD':256,
                '1 TB HDD|256 GB SSD':1000,
                '2 TB SSD':2000,
                '1 TB HDD|512 GB SSD':1000,
                '1 TB HDD|1 TB SSD':1000,
                '512 GB HDD|512 GB SSD':512,
                '1 TB SSD': 1000}
data['SSD_Capacity'] = data['SSD_Capacity'].map(capacity_map)
```

```python
from sklearn.impute import SimpleImputer
```

```python
#We Impute the most frequent value in Review.(As this was non numerical)
imputer1 = SimpleImputer(missing_values=np.NaN,strategy='most_frequent')
data['Review']=imputer1.fit_transform(data['Review'].values.reshape(-1,1))


#We Impute the  Display_size using strategy='constant',fill_value='missing',which creates a 
#completely new category.
imputer2 = SimpleImputer(missing_values=None,strategy='constant',fill_value='missing_value')
data['Display_size']=imputer2.fit_transform(data['Display_size'].values.reshape(-1,1))


#We Impute the median value for Rating and SSD_Capacity. (As this is Numreical and Median never got affected by any outliers.)
imputer3= SimpleImputer(missing_values=np.NaN,strategy='median')
data['Rating']=imputer3.fit_transform(data['Rating'].values.reshape(-1,1))
data['SSD_Capacity']=imputer3.fit_transform(data['SSD_Capacity'].values.reshape(-1,1))
data['MRP']=imputer3.fit_transform(data['MRP'].values.reshape(-1,1))
data['Discount_%']=imputer3.fit_transform(data['Discount_%'].values.reshape(-1,1))
```
